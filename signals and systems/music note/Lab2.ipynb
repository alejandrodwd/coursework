3.1. Load your reference audio signal and plot some periods (5-10) where the amplitude is stable. For instance see Ex. 2.3 from Lab 1.

```python
import util
data, sr = load_audio('./sis1_group103/e1pluck.wav')

t_start = 0.2
t_end = t_start + 7 * T

plot_signals(data, sr=sr, t_start=t_start, t_end=t_end)

n_start = int(t_start * sr)
n_end = int(t_end * sr)
segment = data[n_start:n_end]
t = np.arange(n_end - n_start) / sr

A1 = np.max(np.abs(segment))
print(A1)
```

3.2. Now, define a harmonic signal, `y` whose fundamental frequency is the defined in Lab 1. Go step by step adding a new harmonic in each step. Plot both signals (the reference and the synthesized) and try to reproduce the shape of the reference signal.

```python
K = 8  # using 8 harmonics to start
db_vals = np.array([-15.3, -20.7, -25.7, -15, -40.3, -46.3, -25, -30.6, -47.4])  # dB values of each frequency component
A_k = 10 ** ((db_vals - db_vals[0]) / 20) # convert dB to linear scale
y = np.zeros_like(t)

for k in range(1, K+1):
    y += A_k[k-1] * np.cos(2*np.pi*k*f0*t + k*phi - (k-1)*np.pi/2) #add up components

# normalize so max amplitude matches reference (A1)
y = A1 * y / np.max(np.abs(y))

# plot 
plot_signals([np.real(y)], fs, name=['real part'])
```

3.3. Listen to the synthtesis and remark what are the main differences between the reference and synthesis.

```python
duration = 6  # seconds
t = np.arange(0, duration, 1.0/fs)  # new time vector

# recompute harmonic sum over new time vector
y = np.zeros_like(t)
K = 8
for k in range(1, K+1):
    y += A_k[k-1] * np.cos(2*np.pi*k*f0*t + k*phi - (k-1)*np.pi/2)

y = A1 * y / np.max(np.abs(y))

Audio(y, rate=fs)
```
